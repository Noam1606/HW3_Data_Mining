```python
import requests
import json
api_key=""
```

```python
lst_dest=[]  
origin="Tel Aviv"
with open("dests.txt", "r") as destinations:  ##open the text file, and apply the loop for each city. Append all the data in the list.
    for i in destinations:
        url="https://maps.googleapis.com/maps/api/distancematrix/json?destinations=%s&origins=%s&units=imperial&key=%s" % (i.strip(),origin,api_key)
        response=requests.get(url).json()
        lst_dest.append(response)
    
lst_dest
```

```python
long_lat_dest=[]
with open("dests.txt", "r") as destinations: ##open the text file, and apply the loop for each city. Append all the data in the list.
    for i in destinations:
        url="https://maps.googleapis.com/maps/api/geocode/json?address=%s&key=%s" % (i.strip(),api_key)
        response2=requests.get(url).json()
        long_lat_dest.append(response2)
long_lat_dest
```

```python
tar=[]
distances=[]
durations=[]
lat_long=[]
for i in lst_dest:
    distances.append(i["rows"][0]["elements"][0]["distance"]["value"]//1000)
    durations.append(i["rows"][0]["elements"][0]["duration"]["value"]) 
    tar.append(i["destination_addresses"][0])
for j in long_lat_dest:
    lat_long.append(j["results"][0]["geometry"]["location"])
lat=[]
long=[]
for i in lat_long:
    lat.append(i.get("lat"))
    long.append(i.get("lng"))
```

```python
import pandas as pd

my_dict={"Target":tar,"Distance_km":distances,"Duration":durations,"Longitude":long,"Latitude":lat}
df=pd.DataFrame(my_dict)
df["Duration"]=df["Duration"].apply(lambda x: int(x) )  #changing the format of the duration to hours an  in only
hours=df["Duration"].apply(lambda x: x/3600)
minutes=((hours-hours//1)*60//1).astype(int)
hours=df["Duration"].apply(lambda x: x//3600)
df["Duration"]=hours.astype(str)+" hours-"+minutes.astype(str)+" min"


df
```

```python
print("As we can see the 3 furthest cities from Tel Aviv are Amsterdam, Basel and Valetta ")
df.nlargest(3,"Distance_km")
```
